### TRATANDO NULOS E JUSTIFICATIVAS:
- O valor nulo não é usado nos algoritmos, por isso é importante tratá-los.  A forma como são tratados pode afetar a análise e a clusterização que faremos posteriormente com KMeans.
- Alguns algoritmos tratam nulos nativamente, mas outras formas de tratamento seriam:
  - Excluir o valor nulo;
  - Técnicas Estatísticas: moda, média, mediana, desvio padrão preenchendo os valores nulos com valores fixos;
  - Preencher com Algoritmo de Regressão para aprender padrões e prever os valores não preenchidos.
  - Modelo KNN para encontrar vizinhos similares e possíveis valores para preencher;
- O contexto sempre importa para esse tratamento. No nosso caso, as variáveis nulas que aparecem:
  - 1. Preço: Não podemos analisar mercado sem preço. Não serve para uma análise de precificação, pois não podemos inventar preço
        - 1.1: Vamos excluir apenas as linhas onde o price é nulo.
  - 2. Quartos, Camas e Banheiros: pode-se preencher com uma mediana baseada no tipo de quarto ``room_type`` sendo uma decisão ideal com contexto e segura para não faltar informação

```python
dados = dados.dropna(subset=['price'])
dados['price'].unique()
```

```python
colunas_para_preencher = ['bedrooms', 'beds', 'bathrooms']

for coluna in colunas_para_preencher:
    # Calcula a mediana de cada categoria (Entire home/apt, Private room, Shared room, Hotel room)
    mediana_por_tipo = dados.groupby('room_type')[coluna].transform('median')
    dados[coluna] = dados[coluna].fillna(mediana_por_tipo)

# Verificando se ainda há nulos
print(dados.isnull().sum())
```
